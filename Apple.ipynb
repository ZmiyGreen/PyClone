```python
def simple_step(df: pd.DataFrame, new_name: str) -> pd.DataFrame:
    return  pd.concat([df,
                pd.DataFrame({new_name : [abs((i-j)/j) for i, j in zip(df.CLOSE[1:], df.CLOSE[:-1])]})], 
            axis=1)[:-1]
```
